2. Algunos diccionarios :)

```python
SA = SA[['Name','Family_Name','Glottocode']]
```

```python
SA
```

```python
SA = SA.dropna()
```

```python
SA
```

```python
## familias!

familias = list(SA['Family_Name'])
```

```python
len(familias)
```

```python
len(list(set(familias)))
```

```python
## ¿Cuántas lenguas hay por familia?
```

```python
## creamos un diccionario nombre-familia

nombres = list(SA['Name'])
familias = list(SA['Family_Name'])

name_family = {}

for i in range(len(nombres)):
    nombre = nombres[i]
    familia = familias[i]
    name_family[nombre]=familia
```

```python
name_family
```

```python
## juntemos ahora las lenguas que pertenecen a la misma familia

dict_familia_list = {familia:[] for familia in set(familias)}
```

```python
## recorremos el diccionario name_family y vamos agregando

for lengua in name_family.keys():
    familia = name_family[lengua]
    dict_familia_list[familia]+=[lengua]
```

```python
dict_familia_list['Quechuan']
```

```python
for familia in dict_familia_list.keys():
    print(familia,len(dict_familia_list[familia]))
```

```python
## otra forma! 

from collections import Counter

frecuencias_familias = Counter(familias).most_common(len(familias))
```

```python
frecuencias_familias
```

3. Otra base de datos: [Sails](https://sails.clld.org/)

```python
## https://sails.clld.org/parameters/ARGEX5-1#5/1.746/289.565

argex5_1 = pd.read_csv('argex5-1.txt', sep='\t')  
```

```python
argex5_1
```

```python
## creamos un diccionario nombre-valor

nombres = list(argex5_1['name'])
valores = list(argex5_1['value'])

nombre_valor_1 = {}

for i in range(len(nombres)):
    nombre = nombres[i]
    valor = valores[i]
    nombre_valor_1[nombre]=valor
```

```python
nombre_valor_1
```

```python
## https://sails.clld.org/parameters/ARGEX5-3#5/-13.479/298.401

argex5_3 = pd.read_csv('argex5-3.txt', sep='\t')  
```

```python
argex5_3
```

```python
## creamos un diccionario nombre-valor

nombres = list(argex5_3['name'])
valores = list(argex5_3['value'])

nombre_valor_3 = {}

for i in range(len(nombres)):
    nombre = nombres[i]
    valor = valores[i]
    nombre_valor_3[nombre]=valor
```

```python
nombre_valor_3
```

```python
## juntemos las lenguas que tienen el mismo valor para el rasgos argex5-3

dict_argex5_3_list = {valor:[] for valor in set(nombre_valor_3.values())}
```

```python
dict_argex5_3_list
```

```python
## recorremos el diccionario nombre_valor_3 y vamos agregando

for lengua in nombre_valor_3.keys():
    valor = nombre_valor_3[lengua]
    dict_argex5_3_list[valor]+=[lengua]
```
